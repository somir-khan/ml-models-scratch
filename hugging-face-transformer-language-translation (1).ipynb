```python
# Install required libraries
!pip install transformers sentencepiece -q

# Import libraries
import pandas as pd
from transformers import MarianTokenizer, MarianMTModel

# Load dataset
data_path = '/kaggle/input/article-dataset/articles_data.csv'
news_data = pd.read_csv(data_path, on_bad_lines="skip")

# Show data info
print(news_data.info())

# Load Hugging Face Transformer model and tokenizer
model_name = 'Helsinki-NLP/opus-mt-en-fr'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

# Translation functions
def format_text(language_code, texts):
    return [f">>{language_code}<< {text}" for text in texts]

def translate_texts(texts, model, tokenizer, language="fr"):
    formatted_texts = format_text(language, texts)
    translated = model.generate(**tokenizer(formatted_texts, return_tensors="pt", padding=True))
    return [tokenizer.decode(t, skip_special_tokens=True) for t in translated]

# Translate sample texts
sample_texts = news_data['description'].dropna().sample(5, random_state=42).tolist()
translations = translate_texts(sample_texts, model, tokenizer)

# Display translations neatly
for original, translated in zip(sample_texts, translations):
    print(f"\033[1mOriginal:\033[0m {original}\n\033[92mFrench: {translated}\033[0m\n")
```
